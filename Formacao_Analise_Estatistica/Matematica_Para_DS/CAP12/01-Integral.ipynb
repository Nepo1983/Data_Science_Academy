## Intregral Indefinida

A integral indefinida é um dos principais conceitos do cálculo integral e está intimamente ligada à ideia de antiderivação, ou seja, encontrar a função primitiva de uma função dada. 

É importante distinguir a integral indefinida da integral definida. Enquanto a integral indefinida fornece uma família de funções antiderivadas, a integral definida calcula a área entre a curva da função e o eixo x em um intervalo específico.

```python
import sympy as sp

# Define o símbolo x
x = sp.symbols('x')  

# Define a função
f = x**2 + 2*x + 1

print('Função Original:', f)

# Calcula a integral indefinida
integral_f = sp.integrate(f, x)

print('\nAntiderivada da Função Original (Integral Indefinida):', integral_f)
```

```python
# Representação Visual da Integral Indefinida

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Define o símbolo x
x = sp.symbols('x')   

# Define a função original
f = x**2 + 2*x + 1

print('Função Original:', f)

# Calcula a derivada de f em relação a x
derivada_f = sp.diff(f, x)

print('Derivada de f em relação a x', derivada_f)

# Calcula a integral indefinida
integral_f = sp.integrate(f, x)

print('Antiderivada da Função Original (Integral Indefinida):', integral_f)

# Define os valores de x
x_vals = np.linspace(-10, 10, 400)

# Define os valores de f(x), sua integral e sua derivada
f_vals = x_vals**2 + 2*x_vals + 1
integral_f_vals = (x_vals**3)/3 + x_vals**2 + x_vals
derivada_f_vals = 2*x_vals + 2

# Plota a função f(x), sua derivada e sua integral
plt.figure(figsize=(10,6))
plt.plot(x_vals, f_vals, label = 'f(x) = $x^2 + 2x +1$', color = 'blue')
plt.plot(x_vals, derivada_f_vals, label = "f'(x) = $2x + 2$", color = 'green')
plt.plot(x_vals, integral_f_vals, label = "Integral de f(x)", color = "red")
plt.title('Função f(x), sua Derivada e sua Integral Indefinida')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

Na prática, a antiderivada (ou integral indefinida) de uma função desempenha vários papéis em matemática, física e outras disciplinas científicas. Aqui estão algumas das principais utilizações e significados da antiderivada:

**Acúmulo**: A antiderivada de uma função representa uma quantidade acumulada. Por exemplo, se você tem uma função que representa a taxa de mudança de algo (como velocidade no tempo), sua antiderivada representa a quantidade total acumulada (como a distância percorrida).

**Soluções de Equações Diferenciais**: Equações diferenciais são equações que envolvem derivadas de uma ou mais funções desconhecidas. Encontrar a antiderivada de uma função é o mesmo que resolver uma equação diferencial particularmente simples. Em muitas áreas, desde a física até a biologia e economia, equações diferenciais desempenham um papel central, e a habilidade de encontrar antiderivadas é crucial para resolver tais equações.

**Áreas sob Curvas**: Mesmo que o Teorema Fundamental do Cálculo ligue a antiderivada à integral definida (que representa a área sob a curva de uma função), a própria antiderivada pode ser vista como uma função que acumula a área sob a curva de outra função à medida que se move ao longo do eixo x.

**Problemas de Valor Inicial**: Quando temos uma derivada e uma condição inicial (um valor específico da função em um ponto específico), podemos usar a antiderivada para determinar a função original.

**Cálculo em Física**: Em física, muitas quantidades são definidas em termos de taxas de mudança. Por exemplo, a velocidade é a taxa de mudança da posição. A antiderivada nos permite recuperar a quantidade original (neste caso, a posição) a partir de sua taxa de mudança.

**Aplicações em Engenharia**: Em engenharia, antiderivadas são frequentemente usadas em problemas de análise de sistemas, como determinar a resposta de um sistema elétrico a um determinado input, ou entender o comportamento de sistemas mecânicos sob certas forças.

**Probabilidade e Estatística**: A antiderivada de uma função de densidade de probabilidade nos dá a função de distribuição cumulativa, que é usada para calcular probabilidades. Veremos isso no Estudo de Caso 7.

Esses são apenas alguns exemplos. Em resumo, a antiderivada, ou integral indefinida, é uma ferramenta poderosa com uma ampla variedade de aplicações em matemática e ciências aplicadas. Ela nos permite "reverter" o processo de diferenciação e encontrar a função original a partir de sua derivada, com implicações profundas em várias áreas do conhecimento.

## Propriedades das Integrais Indefinidas

As integrais indefinidas têm várias propriedades importantes que seguem diretamente da definição de integral como uma antiderivada e das propriedades das derivadas. Aqui estão algumas das propriedades mais importantes:

**Linearidade**: Se você tem duas funções, f(x) e g(x), e dois números reais, a e b, então a integral da combinação linear é a combinação linear das integrais, ou seja, ∫[af(x) + bg(x)] dx = a∫f(x) dx + b∫g(x) dx.

**Aditividade**: A integral da soma de duas funções é a soma das integrais das funções. Em outras palavras, ∫[f(x) + g(x)] dx = ∫f(x) dx + ∫g(x) dx.

**Multiplicação Por Uma Constante**: A integral de uma função multiplicada por uma constante é igual à constante multiplicada pela integral da função. Isso é, ∫cf(x) dx = c∫f(x) dx, onde c é uma constante.

**Integral da Função Identidade**: A integral de x com respeito a x é x^2/2 + C, onde C é a constante de integração.

**Integração de Potências de x**: Para qualquer número real n diferente de -1, a integral de x^n com respeito a x é (x^(n+1))/(n+1) + C. Para n = -1, a integral é ln|x| + C.

**Regra da Cadeia Inversa**: Se f é uma função contínua e g é uma função que tem uma antiderivada F, então a integral de f(g(x))*g'(x) dx é F(g(x)) + C.

**Integração Por Partes**: Esta é uma técnica para integrar o produto de duas funções. É a regra do produto para derivadas, mas ao contrário. A fórmula é ∫uv dx = u∫v dx - ∫[u'*(∫v dx)] dx, onde u e v são funções de x e u' é a derivada de u.

Lembrando que, para todas estas propriedades, C representa a constante de integração, que é uma constante arbitrária.

## Intregral Definida

A integral definida é um conceito fundamental do cálculo integral e representa a "área acumulada" sob uma curva em um intervalo específico. Esta "área" pode ser interpretada de diversas maneiras, dependendo do contexto, mas a interpretação geométrica mais comum é a área entre a curva da função e o eixo x no intervalo dado.

```python
import sympy as sp

# Define o símbolo x
x = sp.symbols('x') 

# Define a função
f = x**2 + 2*x + 1

print('Função Original:', f)

# O resultado desse cálculo é o valor numérico da área entre a curva f(x) e o eixo x, no intervalo de 0 a 2.
integral_f = sp.integrate(f, (x, 0, 2))

print('nIntegral Definida:', integral_f)
```

```python
# Representação visual

import matplotlib.pyplot as plt
import numpy as np

# Define os valores de x
x_vals = np.linspace(-1, 3, 400)
f_vals = x_vals**2 + 2*x_vals + 1

# Valores para a área destacada (integral definida)
x_fill = np.linspace(0, 2, 100)
f_fill = x_fill**2 + 2*x_fill + 1

# Plota a função e a área sob a curva
plt.figure(figsize=(10,6))
plt.plot(x_vals, f_vals, label='f(x) = $x^2 + 2x + 1$', color = 'blue')
plt.fill_between(x_fill, f_fill, color='skyblue', label='Área sob a curva entre 0 e 2')
plt.title('Função f(x) e sua Integral Definida de 0 á 2 ')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

## Significado Geométrico da Integral Definida

A integral definida de uma função tem um significado geométrico muito importante: ela representa a área sob a curva da função entre dois pontos no eixo x (os limites de integração).

É importante notar que essa "área" pode ser negativa se a função passar abaixo do eixo x. Em outras palavras, a integral leva em conta não apenas a magnitude da função, mas também a direção. A "área" acima do eixo x contribui positivamente para a integral, enquanto a "área" abaixo do eixo x contribui negativamente.

Se a função for positiva em todo o intervalo de integração, então a integral definida realmente será a área entre a curva da função, o eixo x, e as linhas verticais nos pontos de início e fim do intervalo de integração.

Além disso, a integral definida também tem uma interpretação física em muitos contextos. Por exemplo, se uma função representa a taxa de mudança de alguma quantidade em relação ao tempo, a integral da função em um intervalo de tempo dá a mudança total na quantidade durante aquele intervalo.

```python
# Significado Geométrico da Integral Definida

import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt

# Define a função
def f(x):
    return x**2

# Calcula a integral definida de 0 a 2
integral, error = quad(f, 0, 2)
print(f"A integral de f(x) de 0 a 2 é: ", {integral})

# Visualização
x = np.linspace(0,2, 1000)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y, 'r', linewidth=2)
ax.fill_between(x, y, color='pink', alpha=0.3)
ax.text(0.5, 0.5, f'Área = {integral:.2f}',
        verticalalignment = 'center',
        horizontalalignment = 'center',
        fontsize=15,
        bbox=dict(boxstyle = 'round', facecolor = 'white', edgecolor = 'red'))
        
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
plt.show()
```

## A Integral Como Limite de Uma Soma

A ideia de uma integral como o limite de uma soma é a base para a definição formal do cálculo integral, geralmente chamado de integral de Riemann.

Aqui está a ideia básica:

Suponha que você tenha uma função f(x) e queira encontrar a área sob a curva desta função, entre dois pontos a e b no eixo x. Uma maneira de estimar essa área é dividir o intervalo [a, b] em pequenos subintervalos, encontrar a altura de cada subintervalo usando a função f, multiplicar essa altura pela largura do subintervalo para obter a área de um retângulo, e então somar todas essas áreas de retângulo.

A soma dessas áreas de retângulo é chamada de soma de Riemann. Cada retângulo fornece uma aproximação para a área sob a curva sobre o subintervalo correspondente, então ao somar todas essas áreas de retângulo juntas, obtemos uma aproximação para a área total sob a curva.

Mas ainda é apenas uma aproximação. Para obter a área exata, precisamos considerar o limite à medida que o número de subintervalos vai para o infinito (e, portanto, a largura de cada subintervalo vai para zero). Esse limite das somas de Riemann é o que chamamos de integral de f(x) de a a b, e é denotado ∫_a^b f(x) dx.

A definição formal de uma integral de Riemann é um pouco mais complicada, porque temos que considerar não apenas a divisão do intervalo [a, b] em subintervalos, mas também a escolha dos pontos específicos dentro de cada subintervalo que usamos para calcular as alturas dos retângulos. Mas a ideia básica é a mesma: uma integral é o limite de uma soma de áreas de retângulo à medida que a largura dos retângulos vai para zero.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a função
def f(x):
    return x**2

# Define a soma de Riemann
def riemann_sum(f, a, b, n):
    width = float(b - a) / n
    return sum(f(a + i*width)*width for i in range(n))

# Calcula a soma de Riemann com diferentes números de retângulos
for n in [10, 100, 1000, 10000]:
    result = riemann_sum(f, 0, 1, n)
    print(f"Riemann sum for n = {n}: {result}")

# Visualização da soma de Riemann para n = 10
x = np.linspace(0, 1, 1000)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y, 'r', linewidth=2)

n = 10
width = 1.0 / n
xi = np.linspace(0, 1, n)
yi = f(xi)
ax.bar(xi, yi, width=width, align='edge', alpha=0.5)

ax.set_xlabel('x')
ax.set_ylabel('f(x)')
plt.show()
```

## Exemplo de Aplicação do Cálculo de Integral na Regressão Logística

A Regressão Logística é uma técnica estatística e de aprendizado de máquina usada para modelar a probabilidade de uma variável dependente categórica binária. Em outras palavras, é usada quando a variável de saída (ou resposta) é binária, como "sim" ou "não", "1" ou "0", "verdadeiro" ou "falso", e assim por diante.

Aqui estão os pontos-chave da Regressão Logística:

**Função Logística/Sigmoide**: A Regressão Logística utiliza a função sigmoide (também chamada de função logística) para "espremer" os valores de saída entre 0 e 1, o que é útil para representar probabilidades. 

**Modelagem da Probabilidade**: Ao invés de modelar a resposta diretamente (como na regressão linear), a Regressão Logística modela a log-odds da probabilidade da resposta ser "1". A odds é a razão da probabilidade de sucesso para a probabilidade de falha e o logaritmo dessa razão é chamado de log-odds ou logit.

**Estimação de Parâmetros**: Os parâmetros da Regressão Logística são estimados usando um método chamado Maximum Likelihood Estimation (MLE). O objetivo do MLE é encontrar os valores dos parâmetros que maximizam a probabilidade (ou verossimilhança) dos dados observados.

**Predição**: Uma vez que os parâmetros são estimados, podemos usar o modelo para prever a probabilidade da classe positiva. Com base em um limiar de decisão (muitas vezes 0,5), a probabilidade prevista é então classificada em uma das duas classes.

**Multinomial e Regressão Logística Ordinal**: Embora a descrição acima se refira à regressão logística binária, existem extensões do modelo para situações em que a variável de resposta tem mais de duas categorias. Quando as categorias não têm uma ordem natural, usa-se a regressão logística multinomial; quando as categorias têm uma ordem natural (por exemplo, "baixo", "médio", "alto"), usa-se a regressão logística ordinal.

**Aplicações**: A Regressão Logística é amplamente usada em áreas como medicina (para prever a ocorrência de uma doença com base em fatores de risco), finanças (para prever inadimplência de um empréstimo), marketing (para prever se um cliente comprará um produto) e muitos outros campos.

A implementação de Regressão Logística geralmente não envolve o cálculo direto da integral indefinida. Ao invés disso, a estimativa dos coeficientes da Regressão Logística é feita através da maximização da verossimilhança (usando métodos numéricos como o método de Newton-Raphson ou gradiente descendente, por exemplo).

No entanto, podemos pensar na função sigmoide (usada na Regressão Logística) e sua integral, que é a função logística.

A integral indefinida da função logística é uma expressão que não pode ser representada em termos de funções elementares padrão (como polinômios, funções trigonométricas, exponenciais, etc.). Em vez disso, ela é frequentemente expressa em termos da função Logística Inversa ou função Logit. 

A integral é:

![DSA](imagens/form2.png)

```python
import sympy as sp
import numpy as np

# Define a variável e a função logística
x = sp.symbols('x')

# Define a função sigmoide
sigmoid = 1 / (1 + sp.exp(-x))

# Calcula a integral da função sigmoide
integral_sigmoid = sp.integrate(sigmoid, x)

print(integral_sigmoid)
```

```python
# Converte a função integral em uma função lambda Python para previsão e plotagem 

lambda_integral = sp.lambdify(x, integral_sigmoid, "numpy")
```

```python
# Dados fictícios de X
data_x = np.linspace(-10, 10, 400)
```

```python
len(data_x)
```

```python
# Previsão do valor de y
data_y = lambda_integral(data_x)
```

```python
# Plote
plt.figure(figsize=(8,5))
plt.plot(data_x, data_y, label = "Integral da função Sigmoide")
plt.title('Integral da Função Sigmoide como "Probabilidade Acumulada"')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```
