Podemos calcular a distância entre dois pontos A (x1, y1) e B (x2, y2) usando a fórmula Euclidiana.

```python
# Import
import math

# Função
def calcular_distancia(ponto1, ponto2):
    return math.sqrt((ponto2[0] - ponto1[0])**2 + (ponto2[1] - ponto1[1])**2)

# Pontos
ponto1 = [1, 2]
ponto2 = [4, 6]

# Exectuta a função
distancia = calcular_distancia(ponto1, ponto2)

print(f"Distância entre os pontos: {distancia}")
```

## Exemplo 6: Distância Entre 2 Pontos em R³

A noção de distância entre dois pontos no espaço tridimensional (R³) é uma generalização direta da distância euclidiana em um plano bidimensional (R²). No espaço tridimensional, cada ponto é representado por uma tripla ordenada de coordenadas (x, y, z).

A fórmula da distância Euclidiana entre dois pontos em R³ é uma extensão da fórmula em R². 

Como Funciona:

**Diferença de Coordenadas**: Primeiro, calculamos a diferença entre as coordenadas correspondentes dos dois pontos. Isso nos dá três valores: x2 - x1, y2 - y1 e z2 - z1

**Quadrado das Diferenças**: Em seguida, elevamos cada uma dessas diferenças ao quadrado.

**Soma dos Quadrados**: Somamos os quadrados das três diferenças.

**Raiz Quadrada**: Finalmente, tiramos a raiz quadrada da soma para obter a distância euclidiana entre os dois pontos.

```python
# Função para calcular a distância euclidiana
def calcular_distancia_r3(ponto1, ponto2):
    return math.sqrt((ponto2[0] - ponto1[0])**2 + (ponto2[1] - ponto1[1])**2 + (ponto2[2] - ponto1[2])**2)

# Dados
ponto1 = [1, 2, 3]
ponto2 = [4, 5, 6]

# Executa a função
distancia = calcular_distancia_r3(ponto1, ponto2)

print(f"Distância entre os pontos de R³: {distancia}")
```

## Exemplo 10: Curvas de Nível

Curvas de nível ou contornos são gráficos 2D que representam os locais onde a função assume um valor constante.

Curvas de nível são um conceito matemático e gráfico utilizado para representar uma função de duas ou mais variáveis. Em termos mais simples, uma curva de nível é um conjunto de pontos no espaço onde a função assume um valor constante. Para funções de duas variáveis, estas curvas são frequentemente representadas em um gráfico bidimensional, enquanto para funções de três variáveis, as "curvas" são, na verdade, superfícies de nível tridimensional.

Como são Construídas:

Se você tem uma função f(x, y), a curva de nível correspondente a um valor constante C é o conjunto de pontos (x, y) onde f(x, y) = C.

Escolha um Valor Constante: Para criar uma curva de nível, você primeiro escolhe um valor constante que a função pode assumir.

Localize os Pontos: O próximo passo é encontrar todos os pontos (x, y) onde a função é igual a esse valor constante.

Trace a Curva: Finalmente, você conecta esses pontos para formar uma curva, que é a curva de nível para o valor constante escolhido.

Em Python basta usar plt.contour(). 

Para a função f(x, y) = x² + y², a curva de nível para f(x, y) = 1 é o conjunto de pontos que satisfazem a equação x² + y² = 1. Este é simplesmente um círculo de raio 1 centrado na origem.

```python
# Plot de Curvas de Nível com 20 níveis
plt.contour(x, y, z, levels = 20)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Curvas de Nível f(x, y) = x² + y²')
plt.show()
```

## Exemplo 11: Derivadas Para Funções de 2 Variáveis

As derivadas para funções de duas variáveis, comumente conhecidas como derivadas parciais, descrevem como uma função f(x,y) varia em relação a cada uma de suas variáveis independentes, mantendo a outra constante. 

Para cada ponto (x,y) no domínio da função, as derivadas parciais fornecem as taxas instantâneas de mudança da função na direção dos eixos x e y, respectivamente. Essas derivadas são fundamentais em campos como cálculo multivariável, otimização e ciências físicas, pois oferecem insights sobre o comportamento local da função, como inclinação, pontos críticos e curvatura, em um espaço de duas ou mais dimensões. Esse é o componente matemático principal usado no treinamento de modelos de Deep Learning para aplicações de IA.

Para calcular derivadas parciais de funções de duas (ou mais) variáveis em Python, você pode usar a biblioteca SymPy. Abaixo temos um exemplo de como calcular a primeira e a segunda derivadas parciais de uma função f(x,y):

```python
# Import
from sympy import symbols, diff

# Definindo as variáveis e a função
x, y = symbols('x y')
f = x**2 + y**2

# Calculando as primeiras derivadas parciais
partial_x = diff(f, x)
partial_y = diff(f, y)

# Calculando a segunda derivada parcial em relação a x e y
partial_x_y = diff(partial_x, y)

print(f"A função é: {f}")
print(f"A derivada parcial em relação a x é: {partial_x}")
print(f"A derivada parcial em relação a y é: {partial_y}")
print(f"A segunda derivada parcial em relaçã0 a x e y é: {partial_x_y}")
```

Neste exemplo, utilizamos a função diff da biblioteca SymPy para calcular as derivadas. O primeiro argumento é a função que você quer derivar e o segundo argumento é a variável em relação à qual você quer derivar a função. Você pode usar esta abordagem para funções de qualquer número de variáveis, apenas fornecendo as variáveis adequadas como argumentos para diff.

## Exemplo 12: Limite e Derivada Parcial

O conceito de limite em cálculo descreve o comportamento de uma função à medida que sua variável independente se aproxima de um determinado valor. 

Em termos mais simples, um limite examina o que acontece com uma função quando a entrada se aproxima de um ponto específico, seja de um lado ou de ambos. 

A derivada parcial, por sua vez, é uma extensão do conceito de derivada para funções de diversas variáveis e mede a taxa instantânea de mudança da função em relação a uma das variáveis, mantendo todas as outras constantes. Ela é calculada usando os limites e fornece informações vitais sobre a "inclinação" da superfície da função em relação a uma variável específica em um dado ponto. 

Ambos os conceitos são pilares do cálculo e têm aplicações vastas em várias disciplinas, incluindo física, engenharia, economia e ciências da computação.

Você pode calcular limites e derivadas parciais usando o módulo sympy em Python.

```python
from sympy import symbols, diff, limit

# Símbolos
x, y = symbols('x y')

# Função f(x, y) = x^2 + y^2
f = x**2 + y**2

# Derivada parcial em relação a x
derivada_parcial_x = diff(f, x)

# Derivada parcial em relação a y
derivada_parcial_y = diff(f, y)

# Calculando o limite da derivada parcial em relação a x quando x se aproxima de 1
limite_derivada_x_em_1 = limit(derivada_parcial_x, x, 1)

# Calculando o limite da derivada parcial em relação a y quando y se aproxima de 1
limite_derivada_y_em_1 = limit(derivada_parcial_y, y, 1)

print(f"Derivada parcial em relação a x: {derivada_parcial_x}")
print(f"Derivada parcial em relação a y: {derivada_parcial_y}")
print(f"Limite da derivada parcial em relação a x como x -> 1: {limite_derivada_x_em_1}")
print(f"Limite da derivada parcial em relação a y como y -> 1: {limite_derivada_y_em_1}")
```

Neste exemplo, utilizamos a função limit para calcular os limites das derivadas parciais à medida que x e y se aproximam do valor 1. Observe que, neste caso específico, as derivadas parciais são constantes, então o limite será o próprio valor da derivada parcial.

## Exemplo 15: Função Composta - Regra da Cadeia

A Regra da Cadeia é um conceito fundamental no cálculo e é usada para encontrar a derivada de uma função composta. 

Uma função composta ocorre quando você tem uma função dentro de outra função, como por exemplo f(g(x)). 

A regra da cadeia fornece um método para calcular a derivada de tal composição. Ela diz que para encontrar a derivada da função composta f(g(x)), você deve primeiro encontrar a derivada de f em relação a g(x) e, em seguida, multiplicar isso pela derivada de g(x) em relação a x. 

Em outras palavras, se você tem uma função composta f(g(x)), a derivada dela é a derivada de f avaliada em g(x) multiplicada pela derivada de g(x). Isso é extremamente útil para encontrar derivadas de funções mais complicadas que são formadas pela composição de funções mais simples.

```python
# Símbolo para cálculos simbólicos
x = Symbol('x')

# Função f
f = Function('f')(x)
print('\nFunção f:', f)

# Função g
g = x**2
print("\nFunção g:", g)

# Função h
h = f**2
print("\nFunção h:", h)

# Função v é uma função composta com a combinação de f e g
v = h.subs(f, g)
print('\nFunção Composta v:', v)

# Derivada da função composta v em relação a x
dv_dx = diff(v, x)

print(f"\nA derivada de v em relação a x é: {dv_dx}")
```

No contexto do pacote sympy para Python, o método .subs(f, g) substitui todas as ocorrências da expressão f na expressão h pelo valor ou expressão g. Esse método é comumente utilizado para realizar substituições simbólicas em expressões matemáticas.

## Exemplo 16: Funções Homogêneas - Teorema de Euler

O Teorema de Euler para funções homogêneas afirma que, para uma função homogênea de grau n, a soma das derivadas parciais de cada variável, multiplicadas pelas próprias variáveis, é igual à função original multiplicada pelo grau de homogeneidade. Funções homogêneas são um tipo especial de funções que exibem um comportamento de escala consistente.

Em termos simples, se você tem uma função f (x, y, ...) que é homogênea de grau n, então x multiplicado pela derivada parcial de f em relação a x, mais y multiplicado pela derivada parcial de f em relação a y, e assim por diante, será igual a n vezes a função original. Este teorema é uma ferramenta valiosa para simplificar problemas em diversas áreas como matemática, engenharia e economia.

```python
# Import
from sympy import simplify

# Função homogênea
f = x**2 + y**2

# Derivadas parciais
df_dx = diff(f, x)
df_dy = diff(f, y)

# Verificação do Teorema de Euler
lhs = 2 * f
rhs = x * df_dx + y * df_dy

# Verificando se ambos os lados são iguais
print("O Teorema de Euler é válido:", simplify(lhs - rhs) == 0)
```

## Exemplo 17: Máximos e Mínimos Para Funções de Duas Variáveis

Máximos e mínimos para funções de duas variáveis referem-se aos pontos onde a função atinge seus valores mais altos ou mais baixos em uma região específica. Assim como no cálculo univariado, esses pontos são frequentemente chamados de "pontos críticos" e são locais onde a taxa de mudança da função é zero em relação a ambas as variáveis, ou onde as derivadas parciais não estão definidas. 

Uma vez identificados os pontos críticos, a matriz Hessiana — uma matriz de segundas derivadas parciais — é frequentemente usada para classificar esses pontos como máximos locais, mínimos locais ou pontos de sela (pontos que são nem máximos nem mínimos). Se a matriz Hessiana em um ponto crítico é definida positiva (todos os seus autovalores são positivos), então a função tem um mínimo local naquele ponto. Se a matriz Hessiana é definida negativa (todos os seus autovalores são negativos), então a função tem um máximo local. Se a matriz Hessiana tem autovalores de sinais mistos, então o ponto é um ponto de sela.

```python
# Imports
from sympy import symbols, diff, solve, Matrix

# Define as variáveis e a função
x, y = symbols('x y')
f =- x**2 + y**2
print(f)

# Calcula as derivadas parciais de primeiro grau
f_x = diff(f, x)
f_y = diff(f, y)

# Encontra os pontos críticos (derivadas parciais = 0)
critical_points = solve([f_x, f_y], (x, y))
print(critical_points)

# Verifica se os pontos críticos estão em uma lista
if not isinstance(critical_points, list):
    critical_points = [critical_points]

# Calcula a matriz Hessiana
H = Matrix([[diff(f_x, x), diff(f_x, y)],
            [diff(f_y, x), diff(f_y, y)]])

# Avalia a matriz Hessiana nos pontos críticos e classifica os pontos
for point in critical_points:
    
    x_val, y_val = point[x], point[y]
    
    H_val = H.subs({x: x_val, y: y_val})
    
    eigenvalues_dict = H_val.eigenvals()
    
    eigenvalues = list(eigenvalues_dict.keys())
    
    if all(val > 0 for val in eigenvalues):
        print(f"O ponto ({x_val}, {y_val}) é um mínimo local.")
    elif all(val < 0 for val in eigenvalues):
        print(f"O ponto ({x_val}, {y_val}) é um máximo local.")
    else:
        print(f"O ponto ({x_val}, {y_val} é um ponto de sela)")
```

Outra alternativa:

```python
# Import
from sympy import Matrix, hessian

# Função e sua matriz Hessiana
H = hessian(f, (x, y))
print(H)

# Para um ponto ser mínimo, a matriz Hessiana deve ser definida positiva
print("Matriz Hessiana:", H)
```

A matriz Hessiana que você forneceu é uma matriz 2x2 com os elementos [2,0;0,2]. Esta é uma matriz diagonal e todos os seus elementos diagonais são positivos. Em termos de análise de máximos e mínimos de uma função de duas variáveis, uma matriz Hessiana diagonal com todos os elementos positivos indica que a função tem um mínimo local no ponto onde essa matriz Hessiana foi calculada (em nosso exemplo, 0,0).

A razão é que uma matriz Hessiana definida positiva (ou seja, uma matriz cujos autovalores são todos positivos) sinaliza que a curvatura da função é "para cima" em todas as direções ao redor do ponto em questão. Isso é consistente com a presença de um mínimo local.

## Exemplo 18: Aplicação - Ajuste de Retas Pelo Método dos Mínimos Quadrados

Considere que a relação entre o número de quartos de uma casa (X) e seu preço (y) seja dado pela função abaixo:

y =  m * X + c

Nosso objetivo é encontrar os valores mínimos dos coeficientes m e c para que essa função seja verdadeira. Isso é o que fazemos ao treinar um modelo de regressão linear, por exemplo.

```python
# Import
from numpy.linalg import lstsq

# Dados de exemplo

# Número de quartos
X = np.array([2, 4, 6, 3])
print(X)

# Preço da casa
Y = np.array([120600, 145900, 157200, 122800])
print(Y)

# Matriz A e vetor Y
A = np.vstack([X, np.ones(len(X))]).T
print(A)

# Encontrando os parâmetros da reta
m, c = lstsq(A, Y, rcond = None)[0]
print(m)
print(c)

# Plot dos dados e da reta ajustada
plt.plot(X, Y, 'o', label = 'Dados originais', markersize = 10)
plt.plot(X, m*X + c, 'r', label = 'Reta Ajustada')
plt.legend()
plt.show()
```

Agora usamos os valores mínimos para prever y:

```python
# Funcão
y = m * X + c

# Usando X = 5 e os coeficientes m e c
y = 9945.714285714288 * 5 + 99328.57142857139

print('Valor previsto da casa:', y)
```

## Estudo de Caso 9: Regra da Cadeia em Modelo de Deep Learning

Definição do Problema:

Vamos considerar um problema de classificação binária, onde o objetivo é classificar pontos em duas categorias diferentes, em nosso caso seria: profissional contratado (sim ou 1) e profissional não contratado (não ou 0). Vamos considerar 2 atributos de entrada, por exemplo duas pontuações em um processo de seleção. Uma amostra de dados teria o seguinte formato:

- Candidato 1 --> score1, score2 = contratado (1)
- Candidato 2 --> score1, score2 = não contratado (0)
- Candidato 3 --> score1, score2 = não contratado (0)
- Candidato 4 --> score1, score2 = não contratado (0)

Solução: Algoritmo de Deep Learning com Uso da Regra da Cadeia

A Regra da Cadeia é fundamental no cálculo do gradiente durante o treinamento de modelos de Machine Learning. Isso é especialmente verdadeiro na propagação para trás (Backpropagation), que é um algoritmo para calcular eficientemente os gradientes da função de perda em relação aos pesos da rede. E por que queremos isso? Porque queremos encontrar os mínimos da função de perda, o que torna o erro o menor possível.

Para esta demonstração, usaremos uma rede neural simples com uma camada oculta.

Passos:

- 1- Gerar dados sintéticos.
- 2- Definir a arquitetura do modelo.
- 3- Implementar forward propagation.
- 4- Implementar backward propagation usando a Regra da Cadeia.
- 5- Treinar o modelo e fazer previsões.

1- Gerar dados sintéticos.

```python
# Import
import numpy as np

# Dados
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]]) # Dados de entrada
y = np.array([[0], [1], [1], [0]]) # Rótulos
```

```python
print(X)
```

```python
print(y)
```

2- Definir a arquitetura do modelo.

```python
# Hiperparâmetros
input_size = 2
hidden_size = 2
output_size = 1
learning_rate = 0.1
```

```python
# Inicializa Pesos e Bias com valores randômicos
W1 = np.random.rand(input_size, hidden_size)
W2 = np.random.rand(hidden_size, output_size)
b1 = np.zeros((1, hidden_size))
b2 = np.zeros((1, output_size))
```

```python
# Funções de Ativação 
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Derivada da função de ativação
def sigmoid_derivative(x):
    return x * (1 - x)
```

3- Implementar forward propagation.

```python
# Função para realizar a propagação para frente (forward propagation) na rede neural
def forward_propagation(X):
    
    # Calcula a entrada da camada oculta como o produto escalar entre a entrada X e os pesos W1, mais o bias b1
    hidden_layer_input = np.dot(X, W1) + b1
    
    # Aplica a função de ativação (neste caso, sigmoid) à entrada da camada oculta para obter a saída dessa camada
    hidden_layer_output = sigmoid(hidden_layer_input)
    
    # Calcula a entrada da camada de saída como o produto escalar entre a saída da camada oculta e os pesos W2, 
    # mais o bias b2
    output_layer_input = np.dot(hidden_layer_output, W2) + b2
    
    # Aplica a função de ativação à entrada da camada de saída para obter a saída final (previsão) da rede
    predicted_output = sigmoid(output_layer_input)
        
    # Retorna a saída da camada oculta e a saída (previsão) da rede neural
    return hidden_layer_output, predicted_output
```

4- Implementar backward propagation usando a Regra da Cadeia.

```python
# Função para calcular o erro do modelo (valor real - valor previsto)
def func_erro(y, predicted_output):
    erro = y - predicted_output
    return erro
```

```python
# Função para realizar a retropropagação (backward propagation) na rede neural
def backward_propagation(y, predicted_output, hidden_layer_output):    
    # Calcula o erro na camada de saída como a diferença entre o valor real (y) e o valor previsto (predicted_output)
    # Usamos a função de custo (função de erro). Nosso objetivo é otimizar esta função encontrando seus valores mínimos
    output_error = func_erro(y, predicted_output)
    
    # Calcula o "delta" da camada de saída, que é o erro ponderado pela derivada da função de 
    # ativação (neste caso, sigmoid)
    output_delta = output_error * sigmoid_derivative(predicted_output)
    
    # Calcula o erro na camada oculta. 
    # O erro na camada de saída (output_delta) é propagado de volta para a camada oculta usando os pesos W2
    hidden_layer_error = output_delta.dot(W2.T)
    
    # Calcula o "delta" da camada oculta, que é o erro ponderado pela derivada da função de 
    # ativação (neste caso, sigmoid)
    hidden_layer_delta = hidden_layer_error * sigmoid_derivative(hidden_layer_output)
    
    # Retorna os "deltas" da camada oculta e da camada de saída, que serão usados para atualizar os pesos e bias
    return hidden_layer_delta, output_delta
```

5- Treinar o modelo e fazer previsões.

```python
# Inicia um loop de treinamento que será executado por 10.000 épocas
for epoch in range(10000):    
    # Realiza a propagação para frente para obter a saída prevista e a saída da camada oculta
    hidden_layer_output, predicted_output = forward_propagation(X)
    
    # Realiza a retropropagação para obter os "deltas" (ou gradientes locais) para as camadas
    hidden_layer_delta, output_delta = backward_propagation(y, predicted_output, hidden_layer_output)
    
    # Atualiza os pesos da camada de saída (W2) usando o "delta" da camada de saída e a saída da camada oculta
    # Multiplica pelo learning_rate para controlar o tamanho do passo durante a atualização
    W2 += hidden_layer_output.T.dot(output_delta) * learning_rate
    
    # Atualiza o bias da camada de saída (b2) usando o "delta" da camada de saída
    # A soma é feita ao longo do eixo 0 para considerar todas as amostras no batch
    b2 += np.sum(output_delta, axis = 0, keepdims = True) * learning_rate
    
    # Atualiza os pesos da camada oculta (W1) usando o "delta" da camada oculta e a entrada original (X)
    # Multiplica pelo learning_rate para controlar o tamanho do passo durante a atualização
    W1 += X.T.dot(hidden_layer_delta) * learning_rate
    
    # Atualiza o bias da camada oculta (b1) usando o "delta" da camada oculta
    # A soma é feita ao longo do eixo 0 para considerar todas as amostras no batch
    b1 += np.sum(hidden_layer_delta, axis = 0, keepdims = True) * learning_rate
```
